```python
import pandas as pd
import numpy as np
import datetime as dt
import requests
import io

url = "http://data.insideairbnb.com/united-states/nc/asheville/2021-12-15/visualisations/listings.csv"
#listingsFile = request.urlopen(url)
listingsFile = requests.get(url)
listings = pd.read_csv(io.StringIO(listingsFile.text), index_col=0)
#listings = pd.read_table(listingsFile)

listings.head()
```

```python
listings.describe()
```

```python
#Calculate the average listing price

#remove free listings (price = 0)
priceFilter = listings['price']==0
prices = listings[~priceFilter]['price']

#calculate the average price
avgPrice = round(prices.mean(),2)

print("The average airbnb price in Asheville is ${} per night.".format(avgPrice))
```

```python
#calculate the price per month for each listing
listings['price_per_month'] = listings['price']*365/12

#calculate the average price per month
avgPricePerMonth = round(listings['price_per_month'].mean(),2)

#compare that average cost of a monthly airbnb to the market rate 
#market rate source: https://www.rentcafe.com/average-rent-market-trends/us/nc/asheville/

ashvilleMarketRate = 1512

print("Average airbnb monthly costs are ${:,.2f}, while the market rate is ${:,.2f}.".format(avgPricePerMonth, ashvilleMarketRate))
```
